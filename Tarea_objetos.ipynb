```python
from abc import ABC, abstractmethod

# ============================================================
# OBJETO 1: SISTEMA DE CANCIONES / REPRODUCTOR
# ============================================================

# 1. Abstracción
class Cancion(ABC):
    def __init__(self, titulo, artista, duracion_seg):
        self.titulo = titulo
        self.__artista = artista          # 2. Encapsulamiento
        self.__duracion = duracion_seg    # 2. Encapsulamiento
        self.__reproducciones = 0         # 2. Encapsulamiento

    @abstractmethod
    def genero(self):
        pass

    # 2. Encapsulamiento — acceso controlado
    def obtener_artista(self):
        return self.__artista

    def obtener_duracion(self):
        minutos = self.__duracion // 60
        segundos = self.__duracion % 60
        return f"{minutos}:{segundos:02d}"

    def obtener_reproducciones(self):
        return self.__reproducciones

    # 4. Polimorfismo (método base con lógica compartida)
    def reproducir(self):
        self.__reproducciones += 1
        print(f"▶ Reproduciendo: '{self.titulo}' — {self.__artista}")
        print(f"  Duración: {self.obtener_duracion()} | Género: {self.genero()}")
        print(f"  Reproducciones: {self.__reproducciones}")

# 3. Herencia
class CancionPop(Cancion):
    def __init__(self, titulo, artista, duracion_seg, tiene_coro_pegajoso):
        super().__init__(titulo, artista, duracion_seg)
        self.tiene_coro_pegajoso = tiene_coro_pegajoso

    def genero(self):           # 4. Polimorfismo
        return "Pop"

# 3. Herencia
class CancionRock(Cancion):
    def __init__(self, titulo, artista, duracion_seg, tiene_guitarra_electrica):
        super().__init__(titulo, artista, duracion_seg)
        self.tiene_guitarra = tiene_guitarra_electrica

    def genero(self):           # 4. Polimorfismo
        return "Rock"

# --- Prueba Objeto 1 ---
print("=" * 50)
print("OBJETO 1: SISTEMA DE CANCIONES")
print("=" * 50)

c1 = CancionPop("Flowers", "Miley Cyrus", 200, True)
c2 = CancionRock("Bohemian Rhapsody", "Queen", 354, True)

c1.reproducir()
c1.reproducir()
print()
c2.reproducir()
print(f"\nArtista de c1: {c1.obtener_artista()}")
print(f"Duración de c2: {c2.obtener_duracion()}")


# ============================================================
# OBJETO 2: SISTEMA DE MASCOTAS / VETERINARIA
# ============================================================

# 1. Abstracción
class Mascota(ABC):
    def __init__(self, nombre, edad, peso_kg):
        self.nombre = nombre
        self.__edad = edad          # 2. Encapsulamiento
        self.__peso = peso_kg       # 2. Encapsulamiento
        self.__vacunas = []         # 2. Encapsulamiento

    @abstractmethod
    def tipo_animal(self):
        pass

    # 2. Encapsulamiento — acceso controlado
    def obtener_edad(self):
        return self.__edad

    def obtener_peso(self):
        return self.__peso

    def obtener_vacunas(self):
        return self.__vacunas

    def vacunar(self, vacuna):
        self.__vacunas.append(vacuna)
        print(f" {self.nombre} fue vacunado/a contra {vacuna}")

    # 4. Polimorfismo
    def mostrar_ficha(self):
        print(f"Mascota: {self.nombre} | Tipo: {self.tipo_animal()}")
        print(f"  Edad: {self.__edad} años | Peso: {self.__peso} kg")
        print(f"  Vacunas: {', '.join(self.__vacunas) if self.__vacunas else 'Ninguna'}")

# 3. Herencia
class Perro(Mascota):
    def __init__(self, nombre, edad, peso_kg, raza):
        super().__init__(nombre, edad, peso_kg)
        self.raza = raza

    def tipo_animal(self):      # 4. Polimorfismo
        return "Perro"

# 3. Herencia
class Gato(Mascota):
    def __init__(self, nombre, edad, peso_kg, es_indoor):
        super().__init__(nombre, edad, peso_kg)
        self.es_indoor = es_indoor

    def tipo_animal(self):      # 4. Polimorfismo
        return "Gato"

# --- Prueba Objeto 2 ---
print()
print("=" * 50)
print("OBJETO 2: SISTEMA DE MASCOTAS")
print("=" * 50)

p = Perro("Simón", 4, 12.5, "biggle")
g = Gato("Luna", 2, 4.2, True)

p.vacunar("Rabia")
p.vacunar("Moquillo")
p.mostrar_ficha()

print()
g.vacunar("Leucemia felina")
g.mostrar_ficha()
print(f"\nEdad de Simón: {p.obtener_edad()} años")
print(f"Peso de Luna: {g.obtener_peso()} kg")


# ============================================================
# OBJETO 3: SISTEMA DE TAREAS / GESTOR
# ============================================================

# 1. Abstracción
class Tarea(ABC):
    def __init__(self, titulo, descripcion, prioridad):
        self.titulo = titulo
        self.__descripcion = descripcion    # 2. Encapsulamiento
        self.__prioridad = prioridad        # 2. Encapsulamiento (1=baja, 3=alta)
        self.__completada = False           # 2. Encapsulamiento

    @abstractmethod
    def tipo_tarea(self):
        pass

    # 2. Encapsulamiento — acceso controlado
    def obtener_descripcion(self):
        return self.__descripcion

    def obtener_prioridad(self):
        niveles = {1: "Baja", 2: "Media", 3: "Alta"}
        return niveles.get(self.__prioridad, "Desconocida")

    def esta_completada(self):
        return self.__completada

    def completar(self):
        self.__completada = True
        print(f"Tarea '{self.titulo}' marcada como completada")

    # 4. Polimorfismo
    def mostrar_tarea(self):
        estado = "Completada" if self.__completada else "Pendiente"
        print(f"[{self.tipo_tarea()}] {self.titulo}")
        print(f"  Descripción: {self.__descripcion}")
        print(f"  Prioridad: {self.obtener_prioridad()} | Estado: {estado}")

# 3. Herencia
class TareaPersonal(Tarea):
    def __init__(self, titulo, descripcion, prioridad, lugar):
        super().__init__(titulo, descripcion, prioridad)
        self.lugar = lugar

    def tipo_tarea(self):       # 4. Polimorfismo
        return "Personal"

# 3. Herencia
class TareaLaboral(Tarea):
    def __init__(self, titulo, descripcion, prioridad, cliente):
        super().__init__(titulo, descripcion, prioridad)
        self.cliente = cliente

    def tipo_tarea(self):       # 4. Polimorfismo
        return "Laboral"

# --- Prueba Objeto 3 ---
print()
print("=" * 50)
print("OBJETO 3: SISTEMA DE TAREAS")
print("=" * 50)

t1 = TareaPersonal("Hacer mercado", "Comprar frutas y verduras", 2, "Éxito")
t2 = TareaLaboral("Entregar informe", "Informe Q1 para el cliente", 3, "Bancolombia")

t1.mostrar_tarea()
print()
t2.mostrar_tarea()

print()
t1.completar()
t2.completar()

print()
t1.mostrar_tarea()
print()
t2.mostrar_tarea()

print(f"\nPrioridad de t1: {t1.obtener_prioridad()}")
print(f"¿t2 completada?: {t2.esta_completada()}")
```
